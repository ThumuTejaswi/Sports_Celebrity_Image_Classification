```python
import shutil
from pathlib import Path

# Source folder on your computer
source = Path(r"C:\Users\Tejaswi\OneDrive\Desktop\project\sports celebrity project\model\images_dataset")

# Destination in Jupyter
destination = Path("model/image_datasets")  # your manually created folder

# Loop over each celebrity folder
for celeb_folder in source.iterdir():
    if celeb_folder.is_dir():
        dest_folder = destination / celeb_folder.name
        dest_folder.mkdir(parents=True, exist_ok=True)  # fixed line
        for img_file in celeb_folder.glob("*.*"):
            shutil.copy(img_file, dest_folder / img_file.name)
```

```python
for folder in destination.iterdir():
    if folder.is_dir():
        print(folder.name, "has", len(list(folder.glob("*.*"))), "images")
```
